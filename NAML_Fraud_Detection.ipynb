### **Naive Bayes**

```python
class NaiveBayes:
  
  def fit(self, X, y):
    self.n_samples, self.n_features = X.shape
    self.classes = np.array([0,1])
    self.n_classes = len(self.classes)
    self.eps = 1e-7

    #find mean, var and priors
    self.mean = np.zeros((self.n_classes, self.n_features))
    self.var = np.zeros((self.n_classes, self.n_features))
    self.priors = np.zeros(self.n_classes)

    for c in self.classes:
      X_filt = X[y == c]
      self.mean[c,:] = np.mean(X_filt, axis=0)
      self.var[c,:] = np.var(X_filt, axis=0)
      #find how often class c occurs in the dataset
      self.priors[c] = X_filt.shape[0] / self.n_samples 

  def predict(self, X):
    y_pred = [self.single_predict(x) for x in X]
    return np.array(y_pred)

  def single_predict(self, x):
    posteriors = list()

    for c in self.classes:
      #P(y)
      prior = np.log(self.priors[c]) 
      #P(X|y)
      class_conditional = np.sum(np.log(self.probability_density_function(c, x) + self.eps)) 
      #P(y|X)
      posterior = prior + class_conditional 

      posteriors.append(posterior)

    #return the index with highest probability
    return self.classes[np.argmax(posteriors)]  

  def probability_density_function(self, c, x): 
      mean = self.mean[c]
      var = self.var[c]
      numerator = np.exp(- (x - mean)**2 / (2 * var))
      denominator = np.sqrt(2 * np.pi * var)
      
      #return P(x_i|y)
      return numerator / denominator
```

### **K Nearest Neighbors**

```python
from collections import Counter
```

```python
class KNN:

  def __init__(self, k=3):
    self.k = k
  
  def fit(self, X, y):
    self.X_train = X
    self.y_train = y

  def predict(self, X):
    y_pred = [self.single_predict(x) for x in X]
    return np.array(y_pred)

  def single_predict(self, x):
    #compute distances from other samples
    distances = [np.linalg.norm(x - x_train) for x_train in self.X_train]

    #get k nearest samples and related labels
    k_nearest_samples = np.argsort(distances)[:self.k]
    k_nearest_labels = [self.y_train[i] for i in k_nearest_samples]

    #retrieve most common label
    most_common = Counter(k_nearest_labels).most_common(1)

    return most_common[0][0]
```

### **Logistic Regression**

```python
class LogisticRegression:

    def __init__(self):
        self.w = None
        self.b = None

    def sigmoid(self, x):
        x = np.clip(x, -16, 16)
        return 1.0 / ( 1.0 + np.exp(-x) )

    def loss(self, X, y):
        num_samples = X.shape[0]
        z = X @ self.w + self.b
        y_pred = self.sigmoid(z)
        return (-1.0/num_samples) * np.sum( y*np.log(y_pred) + (1-y)*np.log(1-y_pred) )

    def fit(self, X, y):
        y = y[:,None]
        num_samples = X.shape[0]
        num_features = X.shape[1]

        # hyperparameters
        num_iterations = 10000
        learning_rate = 0.01

        # initialize parameters
        self.w = np.random.randn(num_features,1) 
        self.b = 0.0

        history_train = [self.loss(X, y)]

        for epoch in range(num_iterations):

            z = X @ self.w + self.b
            y_pred = self.sigmoid(z)

            # compute gradients
            grad_w =  (1.0/num_samples) * ( X.T @ (y_pred-y))
            grad_b = (1.0/num_samples) * np.sum(y_pred-y)

            # update parameters
            self.w -= learning_rate * grad_w
            self.b -= learning_rate * grad_b       

            history_train.append( self.loss(X, y) )

        # print('train loss: %1.3e' %history_train[-1])
        # plt.loglog(history_train, label = 'train loss')


    def predict(self, x):
        z = x @ self.w + self.b
        y_pred = self.sigmoid(z)
        predictions = list()
        for i in range(x.shape[0]):
            if y_pred[i] > 0.5:
                predictions.append(1)
            else: 
                predictions.append(0)
        return np.array(predictions)
```

## **Undersampling applied**

**Train-Test Split Performing Undersampling**

```python
#ALWAYS RUN THIS CELL (both when performing and not performing undersampling)

np.random.seed(0)
data_np = data.to_numpy()
fraud_proportions = [50, 34, 25]

#return the undersampled dataset using the given proportion (fraud : non fraud)
def undersample(dataset, fraud_proportion):
    data_fraud = dataset[dataset[:,-1]==1, :]
    data_good = dataset[dataset[:,-1]==0, :]
    np.random.shuffle(data_fraud)
    np.random.shuffle(data_good)

    num_fraud = data_fraud.shape[0]
    num_good_undersampled = int( (100-fraud_proportion) * (num_fraud/fraud_proportion) )

    data_good_undersampled = data_good[:num_good_undersampled,:]

    # merge good undersampled and fraud, shuffle and return
    dataset_undersampled = np.concatenate((data_good_undersampled, data_fraud)) 
    np.random.shuffle( dataset_undersampled)
    return dataset_undersampled

#split the dataset into training and testing using the given proportion (fraud : no fraud)
def split_dataset(dataset, training_percentage):
    data_fraud = dataset[dataset[:,-1]==1, :]
    data_good = dataset[dataset[:,-1]==0, :]
    np.random.shuffle(data_fraud)
    np.random.shuffle(data_good)

    num_fraud_train = int( data_fraud.shape[0] * training_percentage)
    num_good_train = int( data_good.shape[0] * training_percentage)

    train = np.concatenate((data_fraud[:num_fraud_train,:], data_good[:num_good_train,:]))
    test = np.concatenate((data_fraud[num_fraud_train:,:], data_good[num_good_train:,:]))

    np.random.shuffle( train )
    np.random.shuffle( test )

    return train, test
```

```python
#performing undersampling
x_train_sets, y_train_sets, x_test_sets, y_test_sets = list(), list(), list(), list()
for i in fraud_proportions:
    dataset = undersample(data_np, i)

    train, test = split_dataset(dataset, 0.7)

    x_train_sets.append( train[:,:-1] )
    y_train_sets.append( train[:,-1:] )
    x_test_sets.append( test[:,:-1] )
    y_test_sets.append( test[:,-1:])
```

**Performance Analysis when Undersampling is Applied**

```python
from prettytable import PrettyTable

nb = NaiveBayes()
knn = KNN()
lr = LogisticRegression()

classifiers = [nb, knn, lr]
classifiers_name = ['Naive Bayes', 'K Nearest Neighbors', 'Logistic Regression']
tabels = [PrettyTable(['Metric', '50:50', '66:34', '75:25']) for i in range(len(classifiers))]
labels = ('Accuracy', 'Sensitivity', 'Specificity', 'Precision', 'F-measure', 'Area Under Curve')

training_sets = x_train_sets 
training_labels = y_train_sets

metrics = list()

for i in range(len(classifiers)):

  accuracies = list()
  sensitivities = list()
  specificities = list()
  precisions = list()
  f_measures = list()
  aucs = list()

  for j in range(len(training_sets)):
      
    x_test = x_test_sets[j]
    y_test = y_test_sets[j]

    classifiers[i].fit(training_sets[j], training_labels[j][:,0])
    prediction = classifiers[i].predict(x_test)

    TP = np.sum(np.logical_and(prediction == 1, y_test[:,0] == 1))
    FP = np.sum(np.logical_and(prediction == 1, y_test[:,0] == 0))
    TN = np.sum(np.logical_and(prediction == 0, y_test[:,0] == 0))
    FN = np.sum(np.logical_and(prediction == 0, y_test[:,0] == 1))

    accuracy = (TP + TN) / (TP + FP + TN + FN)
    accuracies.append(accuracy)
    sensitivity = TP / (TP + FN)
    sensitivities.append(sensitivity)
    specificity = TN / (FP + TN)
    specificities.append(specificity)
    precision = TP / (TP + FP)
    precisions.append(precision)
    f_measure = (2 * precision * sensitivity) / (precision + sensitivity)
    f_measures.append(f_measure)
    auc = 0.5 * (sensitivity + specificity)
    aucs.append(auc)
  
  metrics.extend([[accuracies, sensitivities, specificities, precisions, f_measures, aucs]])

  for k in range(len(labels)):
    tabels[i].add_row([labels[k]] + metrics[i][k])

  print(classifiers_name[i])
  print(tabels[i])
```

```python
fig, axs = plt.subplots(2,3, figsize=(24,10))
axs = axs.flatten()

dimensions = ['50:50', '66:34', '75:25']
barWidth = 0.25
pos1 = np.arange(len(dimensions))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]
pos = [pos1, pos2, pos3]

for i in range(len(labels)):
  l = list()
  for j in range(len(classifiers)):
    l.append(metrics[j][i])
  for k in range(len(pos)):
    axs[i].bar(pos[k], l[k], width = barWidth, label = classifiers_name[k])
    axs[i].set_title(labels[i])
    axs[i].set_xticks(pos2)
    axs[i].set_xticklabels(dimensions)
    axs[i].legend()
  
fig.tight_layout()
plt.show()
```
